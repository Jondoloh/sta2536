# Format figures for reports

There are multiple parameters that can be modified in order to format figures. It helps the writer to highlight the key points in a plot, and it also helps the reader to clearly understand the figure. Remember to only change the arguments you want such that we can a) easily read all labels and b) observe the finding.

```python
# load base packages
import numpy as np # Python libraries for numerical ops
import scipy.stats as st # Python libraries for stats
import matplotlib.pyplot as plt # Python libraries for plotting
from scipy.special import gamma # import the gamma function
import matplotlib.pylab as plt # Python libraries for plotting
```

```python
# create objects for the baseline figure
Nsims = 10000
X = np.linspace(0,6,100)
Y1 = -2*np.log(np.random.uniform(0,1,Nsims))
Y2 = Nsims*np.exp(-X/2)/2*(X[1]-X[0])

# create the figure
plt.figure(1234, # ID of the figure
          figsize=(12,5)) # size of the figure (in inches)

plt.hist(Y1, bins=X,
         linewidth=2.5, # width of the lines
         color="skyblue", # color of the bins
         ec="brown", # color of the edges
         log=False, # logarithm scale
         label='Histogram') # label of the curve

plt.plot(X, Y2,
         linewidth=4, # width of the line
         linestyle='dashed', # style of the line
         color="red", # color of the line,
         alpha=0.75, # transparency of the curve
         label='Exponential pdf') # label of the curve

plt.title("Distribution of the exponential random variable\n using a inverse CDF approach", # title
          fontsize=16) # font size 
plt.xlabel(r"Value $x$", # x-axis label
           fontsize=14) # font size
plt.ylabel(r"Frequency", # y-axis label
           fontsize=18) # font size
plt.xticks(np.linspace(0,7,15), # locations of the ticks
           fontsize=15) # font size
plt.yticks(fontsize=20) # font size
plt.legend(loc="upper center", # position of the legend
           fontsize='x-large') #  fontsize of the text in the legend

plt.show() # output the figure in the command prompt
#plt.savefig('NameOfDocument.pdf', transparent=True) # save the figure in a .pdf, .png, etc.
```

We can also modify some format arguments for any figures generated in a notebook/Python file. To do so, we tweak the sizes parameters in the **rcParams.update** function of matplotlib before creating figures. We can still modify parameters for specific figures by using the methodology described earlier.

```python
plt.rcParams.update({'font.size': 16,
                     'legend.fontsize': 'x-large',
                     'figure.figsize': (15, 5),
                     'axes.labelsize': 'x-large',
                     'axes.titlesize':'x-large',
                     'xtick.labelsize':'x-large',
                     'ytick.labelsize':'x-large'})
```

When determining which parameters we can modify, refer to the [matplotlib.pyplot documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html).
